```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np

homedir = 'Citra Tutorial Lab 1/'
```

```python

den1 = io.imread('Citra Tutorial Lab 1/den.jpg')
den2 = 255-den1
plt.figure(figsize=(15,10))
plt.subplot(1,2,1);plt.imshow(den1)
plt.title('Original', fontsize=20); plt.axis("off")
plt.subplot(1,2,2);plt.imshow(den2)
plt.title('Negative', fontsize=20);plt.axis("off")
plt.show()
```

```python
a = den1/255
c1 = 0.5
c2 = 2
l1 = c1*np.log(1+(a))
l2 = c2*np.log(1+(a))

plt.figure(figsize=(21, 14))

plt.subplot(1,3,1);plt.imshow(den1)
plt.title('Original', fontsize=20);plt.axis("off")
plt.subplot(1,3,2); plt.imshow(l1)
plt.title('C = 0.5', fontsize=20); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(l2)
plt.title('C = 2', fontsize=20); plt.axis("off")
plt.show()
                                            
```

```python
a = den1/255
c = 1
p1 = 0.5
p2 = 2
g1 = c*(a**p1)
g2 = c*(a**p2)

plt.figure(figsize=(21, 14))
plt.subplot(1,3,1); plt.imshow(den1)
plt.title('Original', fontsize=20); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(g1)
plt.title('P = 0.5', fontsize=20); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(g2)
plt.title('P = 2', fontsize=20); plt.axis("off")
plt.show()
```

```python
bug = io.imread(homedir+'bug.jpg')
mn = min(bug.flatten())
mx = max(bug.flatten())
b = int(np.floor(255/(mx-mn)))
bug_cs = (bug - mn) * b

plt.subplot(1,2,1); plt.imshow(bug)
plt.title('Original', fontsize=20); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(bug_cs)
plt.title('Contrast Stretching', fontsize=20);
plt.axis("off")
plt.show()
```

```python
from skimage import util
plt.figure(figsize=(15,10))

bug_flat = util.img_as_ubyte(bug).flatten()
plt.subplot(2,1,1)
plt.hist(bug_flat, 256, range=(0,256))
plt.title('Original', fontsize=20)

bug_cs_flat = util.img_as_ubyte(bug_cs).flatten()
plt.subplot(2,1,2)
plt.hist(bug_cs_flat, 256, range=(0,256))
plt.title('Contrast Stretching', fontsize=20)

plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
from skimage import exposure 
bug_eq = exposure.equalize_hist(bug)

plt.figure(figsize=(15, 10))
plt.subplot(1,2,1); plt.imshow(bug)
plt.title('Original', fontsize=20); plt.axis("off")

plt.subplot(1,2,2); plt.imshow(bug_eq)
plt.title('Histogram Equalization', fontsize=20);
plt.axis("off")

plt.show()
```

```python
bug_eq_flat = util.img_as_ubyte(bug_eq).flatten()
plt.subplot(2,1,1)
plt.hist(bug_flat, 256, range=(0,256))
plt.title('Original', fontsize=20)

plt.subplot(2,1,2)
plt.hist(bug_cs_flat, 256, range=(0,256))
plt.title('Contrast Stretching', fontsize=20)

plt.subplots_adjust(hspace=0.3)
plt.show()
```
